```python
import pandas as pd
import json
import requests
```

```python
def get_file_contents(filename):
    """ Given a filename,
        return the contents of that file
    """
    try:
        with open(filename, 'r') as f:
            # It's assumed our file contains a single line,
            # with our API key
            return f.read().strip()
    except FileNotFoundError:
        print("'%s' file not found" % filename)

api_key = get_file_contents('../exploration/lastfm_key')
```

```python
df = pd.read_csv("../exploration/songs.csv")
df['tags_mb'] = df['tags_mb'].str.replace(" ","").str.split(";").transform(lambda x: x[:5])
df['lyrics'] = df['lyrics'].str.lower().str.replace("(","").str.replace(")","").str.split().transform(lambda x: {k:1 for k in  set(x)})
songmbidmap = {}
for artist in df['artist_lastfm'].unique():
        url = f"http://ws.audioscrobbler.com/2.0/?method=artist.gettoptracks&artist={artist}&api_key={api_key}&format=json&limit=40"
        r = requests.get(url = url)
        songs = r.json()['toptracks']['track']
        for song in songs:
            if artist not in songmbidmap:
                songmbidmap[artist] = {}
            songmbidmap[artist][song['name']] = song['mbid'] if 'mbid' in song else ""
```

```python
df['song_mbid'] = ""
def getimage(mbid):
    f"https://coverartarchive.org/release/76df3287-6cda-33eb-8e9a-044b5e15ffdd"
for i, row in df.iterrows():
    df.loc[i]['song_mbid'] = songmbidmap[row['artist_mb']][row['song']]
df.head()
```

```python
cleaned = df[["artist_mb", "tags_mb", "lyrics", "image", "song", "song_mbid"]].rename(columns={"artist_mb": "artist", "tags_mb": "genres", "song": "title"})
size = cleaned.shape
cleaned = cleaned.drop_duplicates(subset=['artist', 'title'])
print(size, cleaned.shape)
cleaned.head()
```
